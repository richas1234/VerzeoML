# Game Data Analysis
Taking a look at the data from app store about vaious strategy games to identify trends between user rating and genre.

### Setup and get Data
Load the data from the csv file and setup other functions.


```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt


GAMES_PATH = os.path.join("datasets", "Games")

def load_game_data(games_path=GAMES_PATH):
    csv_path = os.path.join(games_path, "appstore_games.csv")
    return pd.read_csv(csv_path)
```

```python
games = load_game_data()
games.head()
```

### Data Preparation and cleaning

First of all we are concerned with only a few of the fearures of the data set: Game name, Avrage user rating, User rating count, Price, In-app purchases, Primary genre and genre. So we will drop all the other features of the dataset.

```python
Interested_cols = ['Name', 'Average User Rating', 'User Rating Count', 'Price', 'In-app Purchases', 'Primary Genre', 'Genres']
games = games[Interested_cols]
games.head()
```

```python
games.info()
```

As it can be seen there are 17007 instances of name but other features have less instances. We can't infer anything from incomplete data so we will drop rows having incomplete data like the ones without any rating.

```python
games = games.dropna(axis=0, subset=['User Rating Count'])
games.info()
```

Much better. Now there are very few rows for In-app purchases that is because if the app has no in-app purchases its empty so we will replace null with zero for this feature.

```python
games['In-app Purchases'].fillna(0, inplace=True)
games.info()
```

Now lets look at some number for the various features.

```python
# Category features: Genres
games['Genres'].value_counts()
```

Lets remove some redundant genres from this column like: games, strategy. As we know all the data is for strategy games.

```python
games.replace(to_replace="Games, ", value="", regex=True, inplace=True)
games.replace(to_replace=', Games', value="", regex=True, inplace=True)
games.replace(to_replace="Strategy, ", value="", regex=True, inplace=True)
games.replace(to_replace=", Strategy", value="", regex=True, inplace=True)
games['Genres'].value_counts()
```

```python
games['Primary Genre'].value_counts()
```

As seen above there are a lot off rows with the primary genre being games which is redundant so we are now replacing all the rows with primary genre being games with the first genre of the genres column. (This is based on the assumption that all the genres mentioned in the column Genres are sorted based on the most appropriate to least).

```python
games['Primary Genre'] = games['Primary Genre'].replace(to_replace="Games", value=games['Genres'].replace(regex=[',(.)*'], value=''))
print(games.head())
games['Primary Genre'].value_counts()
```

### Data Analysis
Now that the data is clean we can go ahead and try to understand it.

```python
# Start with summary of all the numeric features
games.describe()
```

```python
# lets plot and see what genres dominate this industry
genre = pd.DataFrame(games['Primary Genre'].value_counts())
genre = genre.reset_index()
genre.columns = ['Genre', 'Frequency']
print(len(genre))
genre.head()
```

```python
y_pos = np.arange(len(genre))       
plt.bar(y_pos, genre['Frequency'], align='center', alpha=0.5)
plt.xticks(y_pos, genre['Genre'])
plt.ylabel('Frequency')
plt.title('Genre frequency for games')
plt.xticks(rotation=90)
plt.figure(figsize=(1, 1))
plt.show()
```

As evident by the above bargraph the most significant genre for games is Entertainment.

Now the ratings in the data is not accurate as 100 people may vote for a game which makes its rating 4 and a single 5 rating for a game makes it a 5 star rated game so, I searched for a way to calculated weighted rating. The below formula is used by IMDB for its ranking.

weighted rating (WR) = (v ÷ (v+m)) × R + (m ÷ (v+m)) × C 

where:
R = average for the movie (mean) = Avrage User Rating.

v = number of votes for the movie = User Rating Count.

m = minimum votes required to be listed in the Top 250 = Assuming the minimum number of review rating required is 100.

C = the mean vote across the whole report = The games.describe() says the mean is 4.06.

```python
# Creating a function to calculate the weighted ratings
def wRating(R, v, m, C=4.06):
    return round(((v / (v+m)) * R + (m / (v+m)) * C), 1)

games['Weighted Ratings'] = wRating(games['Average User Rating'], games['User Rating Count'], 100, 4)
games.head()
```

Now lets take a look at the values according to each genre. (Weighted Rating of a genre is just the average of all the Weighted Ratings recieved by the perticular genre.)

```python
genre_rating = games.groupby(['Primary Genre']).mean()
genre_rating = genre_rating.reset_index()
genre_rating.columns = ['Genre','Average User Rating', 'User Rating Count', 'Price', 'Weighted Ratings']
genre_rating
```

```python
y_pos = np.arange(len(genre_rating))       
plt.bar(y_pos, genre_rating['Weighted Ratings'], align='center', alpha=0.5)
plt.xticks(y_pos, genre_rating['Genre'])
plt.ylabel('Ratings (Weighted)')
plt.title('Rating distribution for game genres')
plt.xticks(rotation=90)
plt.figure(figsize=(1, 1))
plt.show()
```

Now let's identify some of the top rated genres. Output: Magazines & Newspapers, Shopping, Health & Fitness, Photo & Video, Role Playing, etc..

```python
genre_rating.sort_values(by='Weighted Ratings', ascending=False)
```

Now Let's look at the trend od user ratings based on pricing. Via a Graph

```python
genre_rating = genre_rating.sort_values(by='Price')
y_pos = np.arange(len(genre_rating))       
plt.bar(y_pos, genre_rating['Weighted Ratings'], align='center', alpha=0.5)
plt.xticks(y_pos, round(genre_rating['Price'],1))
plt.ylabel('Ratings (Weighted)')
plt.xlabel('Price')
plt.title('Rating distribution based on game prices')
plt.xticks(rotation=90)
plt.figure(figsize=(1, 1))
plt.show()
```
